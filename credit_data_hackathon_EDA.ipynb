QUESTION:- Find the relationship between income & credit amount

```python
Q1 = data['AMT_INCOME_TOTAL'].quantile(0.25)       #removing the outliers for better understanding
Q3 = data['AMT_INCOME_TOTAL'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the DataFrame
data = data[(data['AMT_INCOME_TOTAL'] >= lower_bound) & (data['AMT_INCOME_TOTAL'] <= upper_bound)]
```

```python
plt.figure(figsize=(10, 6))
sb.scatterplot(x='AMT_INCOME_TOTAL', y='AMT_CREDIT', data=data)
plt.title('Income vs Credit Amount')
plt.xlabel('Income')
plt.ylabel('Credit Amount')
plt.show()

```

```python
#INTERPRETATION :- It can be found that the applicants having income range between 50000 -250000 & credit score below 1.5
#                   are more in number.
```

```python
QUESTION:- Analyse the relationship between income & education levels.
```

```python
plt.figure(figsize=(10, 6))
sb.barplot(x='GENDER', y='AMT_INCOME_TOTAL', data=data)
plt.title('Income Across Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Income')
plt.xticks(rotation=45)
plt.show()
```

```python
#INTERPRETATION:- Male applicants earn more than female applicants
```

* LEVEL 3 ANALYSIS( MULTIVARIANT )

QUESTION:- Analyse the Income across Education & Gender

```python
plt.figure(figsize=(10, 6))
sb.boxplot(x='NAME_EDUCATION_TYPE', y='AMT_INCOME_TOTAL', hue='GENDER', data=data)
plt.title("Income Across Education and Gender")
plt.xlabel("Education Level")
plt.ylabel("Income")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.show()
```

```python
#INTERPRETATION :- The interquartile range of income is highest for females who completed academic degree.
 #                   It can also be observed that other than acdemic degree, people who pursue any other type of education 
#                    the salary for male is greater than female
#                  Highest no. of outliers exist among females who have not completed their higher studies & females who pursued
#                   secondary.
```
