```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import json
import seaborn as sb
```

```python
df=pd.read_csv('DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv')
df[:10]
```

```python

df['genres'] = df['genres'].apply(lambda x: ', '.join([item['name'] for item in json.loads(x)]))
df['keywords'] = df['keywords'].apply(lambda x: ', '.join([item['name'] for item in json.loads(x)]))
df['production_companies'] = df['production_companies'].apply(lambda x: ', '.join([item['name'] for item in json.loads(x)]))
df['production_countries'] = df['production_countries'].apply(lambda x: ', '.join([item['name'] for item in json.loads(x)]))
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: ', '.join([item['name'] for item in json.loads(x)]))

```
